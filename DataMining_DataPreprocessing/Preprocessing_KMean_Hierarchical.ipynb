```python
# Thêm các thư viện cần thiết: 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Đọc dữ liệu: 
df = pd.read_csv(r"D:\DTS Health.csv")
# Xem kích thước dữ liệu ban đầu: 
print("\nKích thước dữ liệu:", df.shape)
```

```python
print("Giá trị null: ")
print(df.isnull().sum())
```

```python
# Loại bỏ các dòng chứa giá trị null
df = df.dropna()
print(f"Số lượng dòng còn lại sau khi loại bỏ null: {df.shape[0]}")
```

```python
print("Giá trị trùng lặp:")
print(df.duplicated().sum())
```

```python
# Xuất dữ liệu ra file CSV mới
df.to_csv(r"D:\DTS Health new.csv", index=False)
```

```python
X = df[["SoGioTapTD", "SoLitNuoc"]]
wcss = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Vẽ biểu đồ khuỷu tay
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss) 
plt.xlabel('Số cụm (k)')
plt.ylabel('Tổng bình phương sai số')
plt.title('Biểu đồ khuỷu tay')
plt.grid(True)
plt.show()
```
